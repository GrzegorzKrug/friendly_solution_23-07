```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math
import os
```

```python
import keras
```

```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, Input, Flatten, LSTM
```

```python
time_fts = 3
time_window = 4
out_size = 1

inp = Input(( time_window, time_fts))
lay_inp = LSTM(1)(inp)
# print(lay_inp.shape)
flat = Flatten()(lay_inp)
dens = Dense(10, activation='relu')(flat)
dens = Dense(out_size, activation='linear')(dens)

model = Model(inputs=inp, outputs = dens)

model.compile(optimizer='adam', loss='mse')
```

```python
fwd = 3
x = np.arange(10*time_fts*time_window).astype(float)

"========="
# x = x.reshape(-1, time_fts, time_window)
# noise = np.random.random((time_fts-1, time_window))
# x[:, 1:, :] = noise
# y = x[:, 0, -1] + fwd
"========="
x = x.reshape(-1, time_window, time_fts, )
noise = np.random.random((time_window, time_fts-1))
x[:, :, 1:] = noise
y = x[:, -1, 0] + fwd
"========="

# print(x)
# print(y)

y=y.reshape(-1, out_size)
print(x.shape)
print(y.shape)
# x,y
```

```python
# super_y = y
# super_y
```

```python
hist_ob = model.fit(x, y, epochs=4000)
```

```python
pred = model.predict(x)
```

```python
err = (np.abs(pred-y)**2)
pred-y, err, err.sum()
```

```python
hist_ob.history.keys()
```
